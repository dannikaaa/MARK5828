```
# Run this cell to install required software. You may add any other software
# you would like to use
import statsmodels.api as sm
import seaborn as sns
import pandas as pd

from google.colab import files
```

```
# Upload CSV file to Google CoLab
files.upload()
```

```
# Read CSV file into a variable
df = pd.read_csv('Final Regression Data.csv')
```

```
# Decide which feature you would like to use for the independent (X) variable
x_variable = 'followers'

# Decide which feature you would like to use for the dependent (Y) variable
# Hint: Use 'likes' for you Y variable.
y_variable = 'likes'

# Extract relevant columns from CSV and store in the variables below
X = df['followers']
Y = df['likes']
```

```
# Run LR
X = sm.add_constant(X)
model= sm.OLS(Y, X).fit()
predictions = model.predict(X)
regression_result = model.summary()

print(regression_result)
```

```
# To plot a graph of your model, use the following line of code but substitute
# in the names of your X and Y variables and the name of your CSV file
# as strings.
sns.lmplot(x='followers', y='likes', data=df)
```
