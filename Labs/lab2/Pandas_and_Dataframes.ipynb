##### Indexing and Slicing

To access just one column, you can use the shortcut **df["Name of Column"]**.

```
df["Name"]
```

**[df.loc](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html)** is a Pandas functions which allow you to slice the dataframe by **INDEX** and **COLUMN**


The syntax is:

`df.loc[choose_indexes, choose_columns]`

`choose_indexes` and `choose_columns` can be either:

- A list of labels: eg, `["Column 1", "Column 2", "Column n"]`

- A slice item of the syntax: `Start Column:Stop Column`


```
df.loc[:, ["Name", "Type 1", "Type 2"]] # Extract every index (row) of the columns ["Name", "Type 1", "Type 2"]
```

```
df.loc[3:10, :] # Extract indexes 3 to 10 and every column.
```

```
df.loc[[3, 10, 799], ["Name", "Attack", "Defense"]]
```

##### Merging

**[df.merge()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html)** is similar to **[pd.concat()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html)**, with a few subtle differences:

df.merge() | pd.concat()
--|--
Joins two dataframes only | Joins a list of dataframes
Requires two dataframes with a **common column** | Does not need a common index
Will perform a column-wise join | Can perform a column-wise or row-wise join

**[df.merge()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html)** also has special modes of joining:
- **left**: use only keys from left frame
- **right**: use only keys from right frame
- **outer**: use **union** of keys from both frames
- **inner**: use **intersection** of keys from both frames

```
df_names = df.loc[:1, ["#", "Name", "Type 1", "Type 2"]]
df_names = df_names.append(pd.DataFrame({
    "#": [1000, 1001],
    "Name": ["Jack", "Junbum"],
    "Type 1": ["Homo sapiens", "Homo sapiens"],
    "Type 2": ["Ground", "Ground"],
}))
df_stats = df.loc[:1, ["#", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
df_stats = df_stats.append(pd.DataFrame({
    "#": [1002, 1003],
    "Attack": [1000, 10000],
    "Defense": [1000, 10000],
    "Sp. Atk": [1000, 10000],
    "Sp. Def": [1000, 10000],
    "Speed": [1000, 10000],
}))
```

```
df_names
```

```
df_stats
```

```
df_merge_inner = df_names.merge(df_stats, how="inner", on="#")
df_merge_inner
```

```
df_merge_outer = df_names.merge(df_stats, how="outer", on="#")
df_merge_outer
```

```
df_merge_left = df_names.merge(df_stats, how="left", on="#")
df_merge_left
```

##### Dummification

**[pd.get_dummies()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html)** converts categorical variables into dummy/indicator variables.

A **categorical variable** can take on one of a limited number of options. Eg, The column **Student Grade** can take only one of the following options: **{FL, PS, CR, DN, HD}**, so it would be categorical.

The opposite, a **continuous variable** can take an unlimited number of options. (eg, salary).

```
df_dummies_1 = pd.get_dummies(df, columns=["Generation"])
df_dummies_1.head()
```

```
df_dummies_2 = pd.get_dummies(df, columns=["Type 1", "Type 2", "Generation"], drop_first=True) # Dummifying many columns at once, and dropping the first column (to prevent multicollinearity issues).
df_dummies_2.head()
```

##### Dealing with Missing Data

Often when you deal with data, there will be some missing values.

What values are missing in the pokemon dataset?

remove, or fill

```
df.info()
```

To deal with missing data, we have a few options.

- Remove the data
  - Drop the column (use **[df.drop()](https://pandas.pydata.org/pandas-docs/version/0.21/generated/pandas.DataFrame.drop.html)**)
  - Drop the row (use **[df.dropna()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html)**)
  
- Fill it with "artificial" data (**imputation**)
  - Fill value with the mean, mode, etc of the whole column (use **[df.fillna()](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.fillna.html)**)


```
df_drop_name = df.dropna(subset=["Name"]) # Drop rows with missing data in column "Name"
```

```
df_drop_name.info()
```

```
df_drop_all = df.dropna() # Drop rows with missing data across every column.
df_drop_all.info()
```

```
df_drop_col = df.drop(columns=["Type 2"]) # Drop "Type 2" column
df_drop_col.info()
```

```
df.head()
```

```
df["Type 2"].mode()
```

```
df_fill = df.copy()
df_fill["Type 2"] = df_fill["Type 2"].fillna(df_fill["Type 2"].mode()[0]) # Replace missing values in "Type 2" with the most frequent value (ie, "Flying")
df_fill.head()
```

##### Apply

**[df.apply()](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.apply.html)** performs a transformation across the whole column (just like map with lists).

```
def reverse_string(string):
  lower = string.lower()            # Make lowercase
  reverse = lower[::-1]             # Reverse
  capitalize = reverse.capitalize() # Capitalise first letter
  return capitalize
```

```
df_reverse_name = df.copy()
df_reverse_name["Name"] = df_reverse_name["Name"].fillna('')
df_reverse_name["Name"] = df_reverse_name["Name"].apply(reverse_string)
df_reverse_name.head()
```

```
# Does the same thing
df_reverse_name = df.copy()
df_reverse_name["Name"] = df_reverse_name["Name"].fillna('').apply(lambda x: x.lower()[::-1].capitalize())
df_reverse_name.head()
```
