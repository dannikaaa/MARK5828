## c) Creating a new variable (1 mark)

You are interested to see whether your **face_area** in the picture makes an impact on the likes you get. Here, **face_area** is simply defined as the **fraction of the size of the face rectangle compared to the size of the whole picture**. If the face size makes up half the picture, the **face_area** would be 0.5.

**Question**: A function **get_face_area()** has been defined below. You need to complete the contents of the function so that face_area can be calculated from the given inputs.

```
def get_face_area(dim_wid, dim_hgt, face_wid, face_hgt):
  # Write your code here (and fix up the return value) 
  return (face_wid*face_hgt)/(dim_wid*dim_hgt)
```

```
"""YOU DO NOT NEED TO MODIFY THIS"""
def create_face_var(df):
  area_vars = [
      "pic_width",
      "pic_height",
      "face_rectangle_width",
      "face_rectangle_height"
  ]

  df["face_area"] = df[area_vars].apply(lambda x: get_face_area(x[0], x[1], x[2], x[3]), axis=1)
  return df
```

# Part 3. Analyse your Data (5 marks)

You are not required to write any code in the cell below. Just run it.

This function will
- Remove rows containing categories with a less than 1% frequency
- Dummify the categorical columns and remove the column with the smallest average likes for each category.

```
"""YOU DO NOT NEED TO MODIFY THIS"""
def dummify_drop(df):
  cat_cols = [
      "blur_level",     # [low, medium, high]
      "exposure_level", # [underExposure, goodExposure, overExposure]
      "gender",         # [female, male]
      "glasses_type",   # [NoGlasses, ReadingGlasses, Sunglasses, SwimmingGoggles]
      "noise_level",    # [low, medium, high]
      "year_uploaded",
      "month_uploaded",
      "day_uploaded",
      "face_pos",       # [top_left, top_center..., bottom_right]
  ]

  # Removing Low Frequency Categorical Variables
  for cat in cat_cols:
    df = df[df.groupby(cat)[cat].transform('count').ge(len(df) / 100)]

  df_temp = df.copy()
  df = pd.get_dummies(df, columns=cat_cols)

  # Drop Lowest Frequency Category Column [My own way of doing drop_first=True]
  for cat in cat_cols:
    min_cat = str(cat) + "_" + str(df_temp.groupby(cat)["likes"].mean().idxmin())
    print("Dropping: " + min_cat)
    df = df.drop(columns=min_cat)
    
  return df
```

```
# Complete this
y = df_regression.pop('likes')

# Complete this
x = sm.add_constant(df_regression.astype(float))

model = sm.OLS(y,x).fit()

print(model.summary())
```
